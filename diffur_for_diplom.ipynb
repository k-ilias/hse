```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
```

```python
def f(y,t):
    omega = 1.01
    gamma = 0.01
    wj = 1
    a = 0
    y1, y2 = y #вводим имена новых функций
    return [y2,a*math.cos(omega*t)-gamma*y2+wj**2*math.sin(y1)]
```

```python
t = np.linspace(0,10,40)
```

```python
y0 = [0.1,0.2] #начальные условия
```

```python
w = odeint(f, y0, t)
```

```python
w
```

```python
y1 = w[:,0] #вектор значений u
```

```python
y2 = w[:,1] #вектор значение u'
```

```python
plt.plot(t,y1,'r-',linewidth=2, label = "u(t)")
plt.ylabel("u(t)")
plt.xlabel("t")
plt.show()
```

```python
plt.plot(t,y2,'r-',linewidth=2, label = "u(t)")
plt.ylabel("u'(t)")
plt.xlabel("t")
plt.show()
```
