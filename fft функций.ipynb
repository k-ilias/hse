```python
from numpy.fft import fft
import numpy as np
```

```python
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 0.01
x = np.linspace(0.0, N*T, N)
y = np.sin(np.pi*x) + 0.1*np.sin(np.pi*3*x)
y1 = np.sin(np.pi*x)
y2 = -np.sin(np.pi*x) + 0.1*np.sin(np.pi*3*x) + 0.53*np.sin(np.pi*x/3)
yf = fft(y)
yf1 = fft(y1)
yf2 = fft(y2)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
